```python

# Assuming CRASH_DATE is in a format that pandas can recognize as a date
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])

# Extract year from CRASH_DATE
df['YEAR'] = df['CRASH_DATE'].dt.year

# Find the minimum and maximum year
min_year = df['YEAR'].min()
max_year = df['YEAR'].max()

print(f"The dataset covers from {min_year} to {max_year}.")
```

### Constrain dataset from 2016 - 2023

```python
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])

# Filter the dataframe for the years 2016 to 2023
df = df[(df['CRASH_DATE'].dt.year >= 2016) & (df['CRASH_DATE'].dt.year <= 2023)]
```

### Days of week with most accidents

```python
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], format='%m/%d/%Y')

# Extract the day of the week
df['DAY_OF_WEEK'] = df['CRASH_DATE'].dt.day_name()

# Count the occurrences of each day
accident_counts = df['DAY_OF_WEEK'].value_counts()

# Optionally, sort the counts to see the most frequent days for accidents
accident_counts_sorted = accident_counts.sort_values(ascending=False)

# Display the sorted counts
print(accident_counts_sorted)
```

Most accidents happen on Friday. Thed least accidents happen on Sunday.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], format='%m/%d/%Y')

# Extract the day of the week from 'CRASH_DATE'
df['DAY_OF_WEEK'] = df['CRASH_DATE'].dt.day_name()

# Count the occurrences of each day
accident_counts = df['DAY_OF_WEEK'].value_counts()

# Sort the counts in descending order
accident_counts_sorted = accident_counts.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accident_counts_sorted.index, accident_counts_sorted.values, color='blue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Day of the Week')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(accident_counts_sorted)
```

### Crash By Year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], format='%m/%d/%Y')

# Extract the day of the week from 'CRASH_DATE'
df['YEAR'] = df['CRASH_DATE'].dt.year

# Count the occurrences of each day
accident_counts = df['YEAR'].value_counts()

# Sort the counts in descending order
accident_counts_sorted = accident_counts.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accident_counts_sorted.index, accident_counts_sorted.values, color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(accident_counts_sorted)
```

2018 was the year with the most accidents in Chicago (between 2016-23).

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Assuming 'CRASH_MONTH' contains the month number
# Count the occurrences of each month
crash_hour_counts = df['CRASH_MONTH'].value_counts()

# Sort the counts by month index to ensure they are in chronological order
crash_hour_counts_sorted = crash_hour_counts.sort_index()

# Replace month indices with month names using calendar.month_name
crash_hour_counts_sorted.index = [calendar.month_name[month] for month in crash_hour_counts_sorted.index]

# Print the ranked months by crash frequency
print(crash_hour_counts_sorted)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(crash_hour_counts_sorted.index, crash_hour_counts_sorted.values, color='red')
plt.xlabel('Month of Year')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Month')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```
