```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame
# Calculate the counts of each unique value in 'CONTRIBUTING FACTOR VEHICLE 1'
contributing_factors_counts = df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()

# Get the top 10 contributing factors
top_10_factors = contributing_factors_counts.head(10)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_factors.values, y=top_10_factors.index, palette='viridis')
plt.title('Top 10 Contributing Factors for Vehicles in Collisions (NYC)')
plt.xlabel('Count')
plt.ylabel('Contributing Factors')
plt.show()
```

### Crash by year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y')

# Extract the day of the week from 'CRASH_DATE'
df['DAY_OF_WEEK'] = df['CRASH_DATE'].dt.year

# Count the occurrences of each day
accident_counts = df['DAY_OF_WEEK'].value_counts()

# Sort the counts in descending order
accident_counts_sorted = accident_counts.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accident_counts_sorted.index, accident_counts_sorted.values, color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(accident_counts_sorted)
```

2018 was the peak year for accidents in NYC.

### Hour of Day with Most crashes

```python
df['HOUR'] = pd.to_datetime(df['CRASH TIME'], format='%H:%M').dt.hour

# Count the occurrences of each hour
hourly_accidents = df['HOUR'].value_counts()

# Sort the counts in descending order to see the most accident-prone hours
hourly_accidents_sorted = hourly_accidents.sort_values(ascending=False)

# Display the sorted counts
print(hourly_accidents_sorted)
```

4-5 pm is the peak hour for accidents in NYC. Slightly different from the peak hour of 3 pm in Chicago.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH TIME' to an hour format
df['HOUR'] = pd.to_datetime(df['CRASH TIME'], format='%H:%M').dt.hour

# Count the occurrences of each hour
hourly_accidents = df['HOUR'].value_counts()

# Sort the counts in descending order to see the most accident-prone hours
hourly_accidents_sorted = hourly_accidents.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(hourly_accidents_sorted.index, hourly_accidents_sorted.values, color='red')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Hour of the Day')
plt.xticks(hourly_accidents_sorted.index, rotation=45)  # Ensure all hours are labeled
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(hourly_accidents_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y')

# Extract the month from 'CRASH_DATE'
df['MONTH'] = df['CRASH_DATE'].dt.month

# Count the occurrences of each month
accident_counts = df['MONTH'].value_counts()

# Sort the index to ensure the months are in order
accident_counts_sorted = accident_counts.sort_index()

# Replace the month integers with month names
accident_counts_sorted.index = [calendar.month_name[month] for month in accident_counts_sorted.index]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accident_counts_sorted.index, accident_counts_sorted.values, color='blue')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Month of Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(accident_counts_sorted)

```
