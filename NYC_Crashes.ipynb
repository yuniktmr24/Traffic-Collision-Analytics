### Crash by year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y')

# Extract the day of the week from 'CRASH_DATE'
df['DAY_OF_WEEK'] = df['CRASH_DATE'].dt.year

# Count the occurrences of each day
accident_counts = df['DAY_OF_WEEK'].value_counts()

# Sort the counts in descending order
accident_counts_sorted = accident_counts.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accident_counts_sorted.index, accident_counts_sorted.values, color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(accident_counts_sorted)
```

### Hour of Day with Most crashes

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the crash data
# Convert 'CRASH TIME' to an hour format
df['HOUR'] = pd.to_datetime(df['CRASH TIME'], format='%H:%M').dt.hour

# Count the occurrences of each hour
hourly_accidents = df['HOUR'].value_counts()

# Sort the counts in descending order to see the most accident-prone hours
hourly_accidents_sorted = hourly_accidents.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(hourly_accidents_sorted.index, hourly_accidents_sorted.values, color='red')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Hour of the Day')
plt.xticks(hourly_accidents_sorted.index, rotation=45)  # Ensure all hours are labeled
plt.grid(True)
plt.show()

# Display the sorted counts for verification
print(hourly_accidents_sorted)
```
